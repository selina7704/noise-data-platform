```python
import librosa
import numpy as np
import os
import joblib
import pandas as pd
import time

# ğŸ”¹ ì†ŒìŒ ìœ í˜•ë³„ ë°ì‹œë²¨ ê¸°ì¤€ ì„¤ì •
SPL_REFERENCE = 20e-6  # 0dB ê¸°ì¤€ ìŒì••
DB_REFERENCE = {
    "ì°¨ëŸ‰ ê²½ì ": 100,      # ê¸°ì¤€ 100dB @ 1m (êµí†µ ì†ŒìŒ)
    "ì´ë¥œì°¨ ê²½ì ": 100,    # ê¸°ì¤€ 100dB @ 1m (êµí†µ ì†ŒìŒ)
    "ì‚¬ì´ë Œ": 100,         # ê¸°ì¤€ 100dB @ 1m (êµí†µ ì†ŒìŒ)
    "ì°¨ëŸ‰ ì£¼í–‰ìŒ": 90,     # ê¸°ì¤€ 90dB @ 1m (êµí†µ ì†ŒìŒ)
    "ì´ë¥œì°¨ ì£¼í–‰ìŒ": 90,   # ê¸°ì¤€ 90dB @ 1m (êµí†µ ì†ŒìŒ)
    "ê¸°íƒ€ ì†ŒìŒ": 85,      # ê¸°ì¤€ 85dB @ 1m (ê¸°íƒ€ ì†ŒìŒ)
}

# ğŸ”¹ ê±°ë¦¬ & ë°©í–¥ íŒì • í•¨ìˆ˜
def analyze_audio(file_path, model, noise_type):
    try:
        y, sr = librosa.load(file_path, sr=None, mono=False)  # ì˜¤ë””ì˜¤ ë¡œë“œ

        # ğŸ”¹ ëª¨ë…¸/ìŠ¤í…Œë ˆì˜¤ íŒë³„
        is_stereo = len(y.shape) == 2 and y.shape[0] == 2

        # ğŸ”¹ RMS SPL(í‰ê·  ë°ì‹œë²¨) ê³„ì‚°
        if is_stereo:
            left_channel = y[0]
            right_channel = y[1]
            rms_total = np.sqrt(np.mean((left_channel + right_channel) ** 2)) / 2
        else:
            rms_total = np.sqrt(np.mean(y ** 2))

        rms_spl = 20 * np.log10(rms_total / SPL_REFERENCE + 1e-6)  # RMS SPL ë³€í™˜

        # ğŸ”¹ Peak SPL(ìµœê³  ë°ì‹œë²¨) ê³„ì‚°
        peak_amplitude = np.max(np.abs(y))  # ìµœê³  ì§„í­ ê°’ ì°¾ê¸°
        peak_spl = 20 * np.log10(peak_amplitude / SPL_REFERENCE + 1e-6)  # Peak SPL ë³€í™˜

        # ğŸ”¹ ì†ŒìŒ ìœ í˜•ì— ë”°ë¼ Peak SPL ë˜ëŠ” RMS SPL ì‚¬ìš©
        if noise_type in ["ì°¨ëŸ‰ ê²½ì ", "ì´ë¥œì°¨ ê²½ì ", "ì‚¬ì´ë Œ"]:
            spl_used = peak_spl
        else:
            spl_used = rms_spl

        # ğŸ”¹ ê±°ë¦¬ ê³„ì‚°
        db_ref = DB_REFERENCE.get(noise_type, 85)  # ì†ŒìŒ ìœ í˜•ì— ë§ëŠ” ê¸°ì¤€ ë°ì‹œë²¨ ì ìš©
        estimated_distance = 1 * (10 ** ((db_ref - spl_used) / 20))

        # ğŸ”¹ ê±°ë¦¬ ê³„ì‚°ì´ ë¹„ì •ìƒì ìœ¼ë¡œ í¬ê±°ë‚˜ ì‘ìœ¼ë©´ ì ì ˆí•œ ë²”ìœ„ ë‚´ë¡œ ìˆ˜ì •
        if estimated_distance < 0.1:  # ìµœì†Œ 0.1më¡œ ì„¤ì •
            estimated_distance = 0.1
        elif estimated_distance > 1000:  # ìµœëŒ€ 1000më¡œ ì œí•œ
            estimated_distance = 1000
        estimated_distance = round(estimated_distance, 2)

        # ğŸ”¹ ê±°ë¦¬ ì•ŒëŒ ì„¤ì •
        distance_alert = "ì•ŒëŒ ì—†ìŒ"
        if noise_type == "ê¸°íƒ€ ì†ŒìŒ" and estimated_distance <= 10 and spl_used >= 100:  # 100dB ì´ìƒì´ê³  10ë¯¸í„° ì´ë‚´
            distance_alert = "ğŸš¨ 10ë¯¸í„° ì´ë‚´ ìœ„í—˜ ì†ŒìŒ, ì•ŒëŒ í™œì„±í™”"
        elif noise_type in ["ì°¨ëŸ‰ ê²½ì ", "ì´ë¥œì°¨ ê²½ì "] and estimated_distance <= 10:
            distance_alert = "ğŸš¨ 10ë¯¸í„° ì•ŒëŒ í™œì„±í™”"
        elif noise_type == "ì‚¬ì´ë Œ" and estimated_distance <= 20:
            distance_alert = "ğŸš¨ 20ë¯¸í„° ì•ŒëŒ í™œì„±í™”"
        elif noise_type in ["ì°¨ëŸ‰ ì£¼í–‰ìŒ", "ì´ë¥œì°¨ ì£¼í–‰ìŒ"] and estimated_distance <= 3:
            distance_alert = "ğŸš¨ 3ë¯¸í„° ì´ë‚´ ê°ì§€, ì•ŒëŒ í™œì„±í™”"

        # ğŸ”¹ ë°©í–¥ íŒì • (ìŠ¤í…Œë ˆì˜¤ë§Œ)
        if is_stereo:
            rms_left = np.sqrt(np.mean(left_channel ** 2))
            rms_right = np.sqrt(np.mean(right_channel ** 2))
            spl_left = 20 * np.log10(rms_left / SPL_REFERENCE + 1e-6)
            spl_right = 20 * np.log10(rms_right / SPL_REFERENCE + 1e-6)
            db_difference = spl_left - spl_right
            if abs(db_difference) < 1.5:
                direction = "ì¤‘ì•™"
            elif 1.5 <= abs(db_difference) < 3:
                direction = "ì•½ê°„ ì™¼ìª½" if db_difference > 0 else "ì•½ê°„ ì˜¤ë¥¸ìª½"
            else:
                direction = "ì™¼ìª½" if db_difference > 0 else "ì˜¤ë¥¸ìª½"
        else:
            spl_left = spl_right = db_difference = direction = None  # ëª¨ë…¸ì¼ ê²½ìš° ë°©í–¥ ì—†ìŒ

        return file_path, noise_type, spl_used, estimated_distance, distance_alert, spl_left, spl_right, db_difference, direction

    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {file_path} - {e}")
        return file_path, noise_type, None, None, None, None, None, None, None

# ğŸ”¹ ëª¨ë¸ ë¡œë“œ (ì €ì¥ëœ ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°)
model = joblib.load("lgbm_model.joblib")  # ì—¬ê¸°ì— ì €ì¥ëœ ëª¨ë¸ ê²½ë¡œ ì§€ì •

# ğŸ”¹ ë¶„ì„í•  ë°ì´í„°ê°€ ìˆëŠ” í´ë” (Ubuntu í™˜ê²½)
folder_path = "/home/ubuntu/data/etc_noise_data_test_04/ë™ë¬¼/14.ê°œ"

# ğŸ”¹ ë¶„ì„í•  íŠ¹ì • íŒŒì¼ì„ ì§€ì •
file_name = "7.ë™ë¬¼_9156_1.wav"  # ë¶„ì„í•  íŒŒì¼

# íŒŒì¼ ê²½ë¡œ ì§€ì •
file_path = os.path.join(folder_path, file_name)

# âœ… WAV íŒŒì¼ì—ì„œ íŠ¹ì§• ì¶”ì¶œ (ëª¨ë¸ ì…ë ¥ê°’ìœ¼ë¡œ ì‚¬ìš©ë  íŠ¹ì„± ì¶”ì¶œ)
def extract_mfcc(file_path, n_mfcc=50):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfccs, axis=1)

# íŒŒì¼ë¡œë¶€í„° MFCC íŠ¹ì„± ì¶”ì¶œ
features = extract_mfcc(file_path)

# ëª¨ë¸ ì˜ˆì¸¡ (ì†ŒìŒ ìœ í˜• ì˜ˆì¸¡)
noise_type_prediction = model.predict([features])  # ì˜ˆì¸¡ëœ ì†ŒìŒ ìœ í˜•
# ì‹¤ì œ ì†ŒìŒ ìœ í˜•ìœ¼ë¡œ ë³€í™˜ (ìˆ«ì ì¸ë±ìŠ¤ë¥¼ ì›ë˜ ë ˆì´ë¸”ë¡œ)
noise_type_labels = ['ì´ë¥œì°¨ ê²½ì ', 'ì´ë¥œì°¨ ì£¼í–‰ìŒ', 'ì°¨ëŸ‰ ê²½ì ', 'ì°¨ëŸ‰ ì‚¬ì´ë Œ', 'ì°¨ëŸ‰ ì£¼í–‰ìŒ']  # ë ˆì´ë¸” ìˆœì„œ
noise_type = noise_type_labels[noise_type_prediction[0]]  # ì˜ˆì¸¡ëœ ì†ŒìŒ ìœ í˜• (ì²« ë²ˆì§¸ ê²°ê³¼)

# âœ… OOD íƒì§€ (Softmax í™•ë¥  ì˜ˆì¸¡ì„ ì‚¬ìš©í•˜ì—¬ OOD ì²˜ë¦¬)
def predict_with_ood_detection(model, features, threshold=0.8):
    probs = model.predict_proba([features])  # ì˜ˆì¸¡ëœ í™•ë¥ 
    max_prob = np.max(probs)  # ê°€ì¥ ë†’ì€ í™•ë¥ ì„ ì„ íƒ
    if max_prob < threshold:
        return "ê¸°íƒ€ ì†ŒìŒ"  # OOD (ê¸°íƒ€ ì†ŒìŒ)
    return noise_type_labels[np.argmax(probs)]  # í™•ë¥ ì´ ì„ê³„ê°’ì„ ì´ˆê³¼í•˜ë©´ ì†ŒìŒ ìœ í˜• ë°˜í™˜

# ì˜ˆì¸¡ëœ ì†ŒìŒ ìœ í˜•ì´ OODì¸ ê²½ìš° "ê¸°íƒ€ ì†ŒìŒ"ìœ¼ë¡œ ì²˜ë¦¬
predicted_noise_type = predict_with_ood_detection(model, features, threshold=0.8)

# âœ… ê±°ë¦¬ & ë°©í–¥ ë¶„ì„ ì‹¤í–‰
start_time = time.time()  # ë¶„ì„ ì‹œì‘ ì‹œê°„ ê¸°ë¡
result = analyze_audio(file_path, model, predicted_noise_type)
end_time = time.time()  # ë¶„ì„ ì¢…ë£Œ ì‹œê°„ ê¸°ë¡

# ì‹¤í–‰ ì‹œê°„ ì¶œë ¥
execution_time = end_time - start_time  # ì‹¤í–‰ ì‹œê°„ ê³„ì‚°
print(f"\n=== ë¶„ì„ ê²°ê³¼ ===")
print(f"ì†ŒìŒì› ì¢…ë¥˜: {result[1]}")
print(f"ë°ì‹œë²¨: {result[2]:.2f} dB")
print(f"ì¶”ì • ê±°ë¦¬: {result[3]} ë¯¸í„°")
print(f"ì¶”ì • ë°©í–¥: {result[8] if result[8] is not None else 'ì•Œ ìˆ˜ ì—†ìŒ'}")
print(f"ì•ŒëŒ ê¸°ëŠ¥: {result[4]}")
print(f"\nğŸ”¹ ì‹¤í–‰ ì‹œê°„: {execution_time:.4f} ì´ˆ")
```
