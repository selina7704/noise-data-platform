```python
import os
import json
import librosa
import numpy as np
import pandas as pd
from pathlib import Path

# ğŸ”¹ SPL ê¸°ì¤€ê°’ (ì†ŒìŒ ìœ í˜•ë³„ ë°ì‹œë²¨ ê¸°ì¤€)
DB_REFERENCE = {
    "ì°¨ëŸ‰ ê²½ì ": 100,
    "ì´ë¥œì°¨ ê²½ì ": 100,
    "ì‚¬ì´ë Œ": 100,
    "ì°¨ëŸ‰ ì£¼í–‰ìŒ": 90,
    "ì´ë¥œì°¨ ì£¼í–‰ìŒ": 90,
    "ê¸°íƒ€ ì†ŒìŒ": 85,
}

SPL_REFERENCE = 20e-6  # 0dB ê¸°ì¤€ ìŒì••

# ğŸ“ í´ë” ê²½ë¡œ ì„¤ì •
json_folder_car = Path("C:/Users/yi260/OneDrive/ë°”íƒ• í™”ë©´/WAV & JSON ë¹„êµ í´ë”/stereo_folder/1.ìë™ì°¨")
json_folder_bike = Path("C:/Users/yi260/OneDrive/ë°”íƒ• í™”ë©´/WAV & JSON ë¹„êµ í´ë”/stereo_folder/2.ì´ë¥œìë™ì°¨")
wav_folder_car = Path("C:/Users/yi260/OneDrive/ë°”íƒ• í™”ë©´/WAV & JSON ë¹„êµ í´ë”/stereo_folder/1.ìë™ì°¨_stereo")
wav_folder_bike = Path("C:/Users/yi260/OneDrive/ë°”íƒ• í™”ë©´/WAV & JSON ë¹„êµ í´ë”/stereo_folder/2.ì´ë¥œìë™ì°¨_stereo")

# ğŸ“Œ JSON íŒŒì¼ ë§¤ì¹­ì„ ìœ„í•œ ë”•ì…”ë„ˆë¦¬ ìƒì„±
json_data_dict = {}

def load_json_data(json_folder):
    """ JSON ë°ì´í„° ë¡œë“œ """
    for json_file in json_folder.glob("*.json"):
        with open(json_file, "r", encoding="utf-8") as f:
            data = json.load(f)
            file_name = data["audio"]["fileName"].replace(".wav", "")  # JSONì˜ íŒŒì¼ëª… ì¶”ì¶œ
            json_data_dict[file_name] = {
                "decibel": data["annotations"][0]["decibel"],
                "distance": data["environment"]["distance"],
                "direction": data["environment"]["direction"]
            }

# ğŸ”¹ JSON ë°ì´í„° ë¡œë“œ (ìë™ì°¨ + ì´ë¥œì°¨)
load_json_data(json_folder_car)
load_json_data(json_folder_bike)

# ğŸ“Œ ë¶„ì„ ê²°ê³¼ ì €ì¥ ë¦¬ìŠ¤íŠ¸
comparison_results = []

def analyze_wav(file_path, json_data):
    """ WAV íŒŒì¼ ë¶„ì„ ë° JSONê³¼ ë¹„êµ """
    y, sr = librosa.load(file_path, sr=None, mono=False)

    # ğŸ”¹ ìŠ¤í…Œë ˆì˜¤ ì—¬ë¶€ í™•ì¸
    is_stereo = len(y.shape) == 2 and y.shape[0] == 2

    # ğŸ”¹ ë°ì‹œë²¨ ê³„ì‚° (RMS SPL & Peak SPL)
    rms_total = np.sqrt(np.mean(y ** 2))
    peak_amplitude = np.max(np.abs(y))
    rms_spl = 20 * np.log10(rms_total / SPL_REFERENCE + 1e-6)
    peak_spl = 20 * np.log10(peak_amplitude / SPL_REFERENCE + 1e-6)

    # ğŸ”¹ ê±°ë¦¬ ê³„ì‚° (ê°ì‡ ìœ¨ ì ìš©)
    category = json_data["direction"]
    db_ref = DB_REFERENCE.get(category, 85)
    estimated_distance = 1 * (10 ** ((db_ref - rms_spl) / 20))
    estimated_distance = max(0.1, min(estimated_distance, 50))  # 50m ì´ìƒì€ '50m ì´ìƒ'ìœ¼ë¡œ í‘œê¸°

    # ğŸ”¹ ë°©í–¥ íŒë³„ (ì¢Œ/ìš° SPL ì°¨ì´ ê³„ì‚°)
    if is_stereo:
        left_spl = 20 * np.log10(np.sqrt(np.mean(y[0] ** 2)) / SPL_REFERENCE + 1e-6)
        right_spl = 20 * np.log10(np.sqrt(np.mean(y[1] ** 2)) / SPL_REFERENCE + 1e-6)
        direction = "ì™¼ìª½" if left_spl > right_spl else "ì˜¤ë¥¸ìª½"
    else:
        direction = "ì•Œ ìˆ˜ ì—†ìŒ"

    # ğŸ”¹ JSONê³¼ ë¹„êµ
    json_distance = json_data["distance"]
    json_decibel = json_data["decibel"]
    json_direction = json_data["direction"]

    match_distance = abs(float(json_distance.replace("m", "")) - estimated_distance) < 5  # 5m ì´ë‚´ ì°¨ì´ë©´ ë§¤ì¹­
    match_decibel = abs(float(json_decibel) - rms_spl) < 5  # 5dB ì´ë‚´ ì°¨ì´ë©´ ë§¤ì¹­
    match_direction = (json_direction == direction) if is_stereo else "N/A"

    comparison_results.append({
        "wav_file": file_path.name,
        "json_distance": json_distance,
        "wav_predicted_distance": f"{round(estimated_distance, 1)}m",
        "json_decibel": json_decibel,
        "wav_predicted_decibel": round(rms_spl, 1),
        "json_direction": json_direction,
        "wav_predicted_direction": direction,
        "match_distance": match_distance,
        "match_decibel": match_decibel,
        "match_direction": match_direction
    })

# ğŸ”¹ WAV íŒŒì¼ ë¶„ì„ (ìë™ì°¨ + ì´ë¥œì°¨)
for wav_file in wav_folder_car.glob("*.wav"):
    base_name = wav_file.stem.rsplit("_", 1)[0]  # `_1` ì œê±°
    if base_name in json_data_dict:
        analyze_wav(wav_file, json_data_dict[base_name])

for wav_file in wav_folder_bike.glob("*.wav"):
    base_name = wav_file.stem.rsplit("_", 1)[0]
    if base_name in json_data_dict:
        analyze_wav(wav_file, json_data_dict[base_name])

# ğŸ“Š ê²°ê³¼ ë°ì´í„°í”„ë ˆì„ ìƒì„±
comparison_df = pd.DataFrame(comparison_results)

# ğŸ“ CSV íŒŒì¼ë¡œ ì €ì¥ (íŒŒì¼ í™•ì¸ ê°€ëŠ¥)
comparison_df.to_csv("C:/Users/yi260/OneDrive/ë°”íƒ• í™”ë©´/WAV & JSON ë¹„êµ í´ë”/comparison_results.csv", index=False, encoding="utf-8-sig")

# ğŸ“Œ ë°ì´í„°í”„ë ˆì„ ì¶œë ¥
from IPython.display import display
display(comparison_df)
```
